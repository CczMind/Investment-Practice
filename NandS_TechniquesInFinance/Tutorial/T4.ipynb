## Question1: Simulate 500 Brownian Motion Paths

```python
import time
import numpy as np
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

T = 1.0 # time horizon (option maturity)
N = 100 # number of time steps
M = 500 # number of Monte Carlo paths
#-----------Implement your code below---------------
dt = T / N
sqrtdt = np.sqrt(dt)

W = np.zeros([N + 1, M])

for i in np.arange(N):
    W[i + 1, :] = W[i, :] + sqrtdt * norm.rvs(size = M)
#---------------------------------------------------
tgrid = np.linspace(0.0, T, N + 1)
fig, ax = plt.subplots()
ax.plot(tgrid, W)
plt.show()
```

## Question2: Simulate 20 CIR Process

```python
# Parameters of CIR
T = 1.0 # time horizon (option maturity)
N = 100 # number of time steps
M = 20 # number of CIR paths
dt = T/N # time step size
mu = 20 # Mean
sig = 5 # Standard Deviation
k = 0.1
#-----------Implement your code below---------------
S = np.zeros([N + 1, M])
S[0, :] = mu

for i in np.arange(N):
    S[i + 1, :] = S[i, :] + k * (mu - S[i, :]) * dt + sig * np.sqrt(S[i, :]) * np.sqrt(dt) * norm.rvs(size = M)
#---------------------------------------------------
tgrid = np.linspace(0.0,T,N+1)
fig, ax = plt.subplots()
ax.plot(tgrid, S)
plt.show()
```
