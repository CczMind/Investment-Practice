## Example 1: Monte Carlo Simulation for $\pi$

```python
import random

N = 1000

plt.figure(figsize = (4, 4))
for i in range(N):
    
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    
    if x ** 2 + y ** 2 < 1:
        plt.plot(x, y, "r.")
    else:
        plt.plot(x, y, "b.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].scatter(x, y, s=sizes, c=colors)

ax[0].set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
ax[0].set_title('Figure 1')

ax[1].scatter(x, y, s=sizes, c=colors)

ax[1].set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
ax[1].set_title('Figure 2')

fig.suptitle('Two figures in a row')

plt.show()
```

## Example 4: Histogram

## Example 5: Two Curves in One Figure

```python
import numpy as np

x = np.linspace(-np.pi, np.pi, 20)
y1, y2 = np.cos(x), np.sin(x)
```

```python
import matplotlib.pyplot as plt

gcf = plt.figure()
plt.plot(x, y1, marker = "o", linestyle = "--")
plt.plot(x, y2, marker = "d")

plt.xlabel("X Label (unit)")
plt.xticks([-np.pi / 2, np.pi / 2], labels = ["$-\dfrac{\pi}{2}$", "$\dfrac{\pi}{2}$"])
plt.ylabel("Y Label (unit)")
plt.legend(["cos(x)", "sin(x)"], loc = "lower center")
plt.text(-np.pi, 0, "Python for Finance", fontsize = 20)
plt.grid(True)
plt.show()
```

## Example 6: Dual Y-Axis Plot

the twinx() function is used to create a new axes that shares the same x-axis with the current axes but has an independent y-axis. This is useful when you want to overlay two plots with different y-scales but the same x-axis values, allowing you to visualize relationships between two datasets with different units or ranges.

```python
import matplotlib.pyplot as plt

y3 = 100 * y2

fig, ax1 = plt.subplots()
ax1.plot(x, y1, marker = "o", linestyle = "--")
ax1.set_ylabel("Left Y Label (unit)")

ax2 = ax1.twinx()
ax2.plot(x, y3, color = "pink", linestyle = ":", marker = "o")
ax2.set_ylabel("Right Y Label (unit)")
fig.legend(["cos(x)", "sin(x)"], loc = 'upper right')

plt.axvline(x = 0, color = "gray")
plt.xticks([-np.pi / 2, 0, np.pi / 2], labels = ["$-\dfrac{\pi}{2}$", "0", "$\dfrac{\pi}{2}$"])
plt.xlabel("X Label (unit)")
plt.grid(True)
```

## Example 7: Subplots

`plt.subplots()` in Matplotlib creates a figure and a set of subplots within that figure. It's a convenient way to arrange multiple plots in a single figure canvas. 

```python
import numpy as np

# Create some sample data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)
y4 = np.cos(2 * x)

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# Add data to each subplot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Plot 1')
axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Plot 2')
axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Plot 3')
axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Plot 4')

# Add an overall title for the entire figure
fig.suptitle('Subplots Example')

# Adjust layout to prevent overlap of titles and labels
plt.tight_layout()

# Show the figure with subplots
plt.show()
```
