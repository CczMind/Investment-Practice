# 农产品期货 —— 豆粕为例

## 1. 什么是期货？

- **定义**：期货是一种合约，在未来特定时间以约定价格买卖某种资产（大宗商品或金融资产）。  

- **功能**：  
  1. **价格发现**：通过交易所的集中交易，形成公开透明的价格。  
  2. **风险管理（套期保值）**：生产者、贸易商、农户、加工企业可以用期货锁定成本和收益。  
  3. **投资与投机**：投资者利用价格波动获利。  

- **特点**：  
  - **保证金制度**：只需支付合约价值的一小部分作为保证金，就能放大交易规模。  
  - **杠杆效应**：收益和风险同时被放大。  
  - **双向交易**：既可做多（看涨），也可做空（看跌）。  

---

### 期货市场的主要参与者

1. **套期保值者（Hedgers）**  
   - **目标**：降低或锁定风险。  
   - **例子**：  
     - 航空公司担心未来油价上涨，可以通过买入原油期货锁定成本。  
     - 农户担心小麦价格下跌，可以提前卖出小麦期货锁定收益。  

2. **投机者（Speculators）**  
   - **目标**：通过价格波动获取利润。  
   - 不持有实际商品，只关注价格涨跌。  
   - **例子**：  
     - 投资者看空黄金，可以直接卖出黄金期货，从下跌中获利。  
     - 看多股指期货的交易者，通过上涨获得杠杆收益。  

3. **套利者（Arbitrageurs）**  
   - **目标**：利用不同市场或不同合约间的价差获利。  
   - **例子**：  
     - 同一时间，发现国内外大豆期货存在价差，可以低买高卖进行套利。  
     - 在近月和远月合约之间进行跨期套利。  

👉 这三类角色共同构成期货市场的生态：**套保者提供真实需求，投机者提供流动性，套利者维持市场价格的均衡。**



## 2. 什么是农产品期货？

- **定义**：农产品期货是以大豆、玉米、小麦、白糖、棉花、苹果等农产品为标的的期货合约。  

- **功能与意义**：  
  - 帮助农业生产者、加工厂、养殖场管理价格风险；  
  - 与居民日常生活息息相关（粮食、肉类、饮料、衣物等）。  

- **特点**：  
  - **季节性强**：播种与收割周期决定价格节奏。  
  - **受供需与气候政策影响大**：天气灾害、库存水平、进出口关税、政府补贴。  
  - **品种丰富**：豆粕、大豆、玉米、小麦、白糖、苹果等。

### （4）小结 & 交易启发

- **短期**：库存、出口/到港节奏、压榨利润与基差决定波动；  
- **中期**：**拉尼娜/厄尔尼诺** + **播种/收割** + **USDA 报告** → 常触发价格级别的方向选择；  
- **实操提示**：  
  - 基本面决定方向，**技术面择时**（均线/价量/波动率）；  
  - 关注 **跨期/跨品种**价差与季节性因子；  
  - 套保与风险控制（保证金、止损、事件前后仓位管理）。

---
### 名词解释
- **库存**：市场可用存量。低库存 → 货源紧 → 价格易涨；高库存 → 价格受压。  
- **出口/到港节奏**：海外装船与国内到港速度。到港推迟常引发**近端偏紧**。  
- **压榨利润（榨利）**：大豆压出豆油+豆粕的利润。榨利好 → 厂商增产；榨利差 → 开机下降。  
- **基差**：现货价 − 期货价（本课件统一此定义）。基差走强 = 现货更强，偏利多近月。  
- **拉尼娜/厄尔尼诺**：海温异常引发的气候现象。拉尼娜偏干旱（利多美豆/豆粕），厄尔尼诺偏多雨（利空概率升）。  
- **播种/收割**：美 4–5月播/9–10月收；巴西 3–5月收；阿根廷 11–12月播/5月收。常伴随供给与预期波动。  
- **USDA 报告**：美国农业部供需月报（WASDE）与种植/库存数据，库存消费比（Stocks-to-use）越低越偏多。  
- **跨期价差**：同品种不同到期的相对交易（近月−远月），押注**曲线形态变化**。  
- **正价差/逆价差**：远月>近月为正价差（Contango）；近月>远月为逆价差（Backwardation）。  
- **技术面择时**：均线（金叉/死叉）、价量（放量突破/缩量回踩）、波动率（布林带开口、历史波动率变化）等。  
- **套保/保证金/止损/仓位**：企业对冲价格风险；保证金带来杠杆；止损与仓位管理是风控核心。

### 1. 策略背后的假设
- **近端供需更紧**：拉尼娜导致美豆减产 + 全球库消比下降 → 近期现货紧缺。  
- **远端预期尚未完全反映**：远月合约更多反映“未来新作产量”，供给压力可能缓解。  
- **结果**：近月价格涨得更快 → 价差收敛甚至转逆价差 → “多近空远”获利。

---

### 2. 假设成立的条件（策略有效时）
- **气候冲击**：拉尼娜/干旱，导致主产区减产。  
- **到港推迟**：船期延误、港口拥堵、运输受阻 → 短期缺货。  
- **库存偏低**：国内港口、商业库存持续下降。  
- **需求强劲**：猪禽养殖恢复、饲料厂集中补库。  
- **基差走强**：现货升水期货，近端供需紧张。  

---

### 3. 假设不成立的条件（策略失效时）
- **供应突然缓解**：集中到港、抛储、进口配额放开。  
- **需求走弱**：猪病、去产能、饲料降配豆粕。  
- **远月利多更强**：新作利多超过近端 → 远月价格涨得比近月更快。  
- **数据反转**：USDA/气象模型上调产量、库存，或天气改善。  

---

### 4. 风险提示
- 策略依赖的前提若落空（如天气好转、到港超预期），曲线会重新走向正价差 → 多近空远会亏损。  
- 必须设定 **止损条件**：如基差连续转弱、价差重新走阔、远月转强。  

---

### 5. 一句话总结
- **能赚钱的时候**：眼前的货更稀缺（基差↑、近端供需紧张）。  
- **会亏钱的时候**：供给缓解或需求走弱（基差↓、远月更强）。

````python
# 课堂演示：跨期价差（多近空远）的盈亏来源
# 策略假设：拉尼娜+低库消比 → 近期现货紧缺 → 近月价格强于远月
# 注意：若到港增加/库存回升/需求走弱，该假设可能失效 → 策略亏损

import pandas as pd

# 初始价格（元/吨）
near0 = 3600   # 近月
far0  = 3650   # 远月
spread0 = near0 - far0  # 初始价差（近-远），正价差通常为负数

# 事件后价格（假设：近端紧缺 → 近月涨幅大于远月）
near1 = 3780
far1  = 3700
spread1 = near1 - far1

# 结果汇总
df = pd.DataFrame({
    "近月价格": [near0, near1],
    "远月价格": [far0,  far1 ],
    "价差(近-远)": [spread0, spread1]
}, index=["事件前", "事件后"])

# 单手吨位（豆粕常见合约为10吨/手）
TONS_PER_LOT = 10

# 做多价差 = 多近月 + 空远月
pnl_near = (near1 - near0) * TONS_PER_LOT     # 多近月盈利
pnl_far  = (far0  - far1 ) * TONS_PER_LOT     # 空远月盈利
pnl_total = pnl_near + pnl_far

print(df)
print("\n—— 跨期多近空远 PnL（以10吨/手计）——")
print(f"近月盈亏: {pnl_near:,.0f} 元")
print(f"远月盈亏: {pnl_far:,.0f} 元")
print(f"合计盈亏: {pnl_total:,.0f} 元")
print(f"价差变化: {spread0} → {spread1} ；核心赚的是『近端供需更紧』假设成立时的价差收敛/反转")
````

## 4. 技术面与交易策略

技术分析是帮助交易者择时的重要工具。和基本面结合，可以更好地指导交易。

---

### （1）技术指标

- **均线（MA，Moving Average）**
  - **定义**：过去一段时间的平均价格（如5日、20日、60日）。  
  - **积极信号**：
    - 短期均线（如5日）向上突破长期均线（如20日、60日） → **金叉**，趋势看涨。  
    - 价格站在均线之上并均线多头排列 → **上涨趋势确认**。  
  - **消极信号**：
    - 短期均线下穿长期均线 → **死叉**，趋势转弱。  
    - 价格跌破60日均线并均线空头排列 → **下跌趋势确认**。

  - **定义**：由快(DIF)慢(DEA)两条指数均线差值构成，配合红绿柱来观察动能。

    - DIF线下穿DEA线（形成死叉） → 下跌动能增强。  
    - 柱状图由正转负 → 空头能量释放。

- **布林带（Bollinger Bands）**
  - **定义**：以价格均线为中轨，上下轨根据波动率决定。  
  - **积极信号**：
    - 价格突破上轨并伴随放量 → 突破行情，可能进一步上涨。  
    - 中轨获得支撑，价格回踩不破 → 趋势延续。  
  - **消极信号**：
    - 价格跌破下轨并伴随放量 → 下跌趋势加速。  
    - 中轨失守且反抽失败 → 下跌延续。

### （2）资金面

- **交易量（Volume）**
  - 上涨+放量 → 买盘踊跃，多头确认。  
  - 下跌+放量 → 抛压沉重，空头确认。  

- **持仓量（OI, Open Interest）**
  - 上涨+持仓增加 → 新资金流入，趋势可能延续。  
  - 上涨+持仓减少 → 多头平仓，涨势存疑。  
  - 下跌+持仓增加 → 新空头入场，趋势可能延续。  
  - 下跌+持仓减少 → 空头平仓，跌势或见底。

### （3）套利策略

- **跨期套利（Calendar Spread）**
  - 同一品种不同交割月的合约。  
  - **牛市价差策略**：多近月，空远月（近月因供需紧张更强）。  
  - **熊市价差策略**：空近月，多远月（远月因供应宽松更强）。  

- **跨品种套利（Inter-commodity Spread）**
  - 豆油 vs 棕榈油：油脂类相对交易。  
  - 豆粕 vs 玉米/小麦：饲料原料间的替代关系。  

- **跨市场套利（Inter-market Spread）**
  - 国内 vs 国际同品种合约：如大连豆粕 vs 芝加哥豆粕（CBOT）。  
  - 利用价差波动获利。  

---

### （4）案例：2018年中美贸易摩擦

- **背景**：中美贸易战升级，中国对美大豆加征关税 → 大豆进口受限。  
- **基本面**：
  - 美国大豆进口量下降；  
  - 国内豆粕库存减少，饲料企业抢货。  
- **技术面**：
  - 价格突破前期高点阻力位，形成上升趋势。  
  - 均线金叉出现，MACD翻红 → **买入信号**。  
- **结果**：
  - 豆粕价格大幅上涨，套利资金和投机资金同步进场。  

---

### 📌 总结提示

- 技术面提供择时工具，但要结合基本面背景。  
- **积极信号**：金叉、突破上轨、放量上涨、持仓增加。  
- **消极信号**：死叉、跌破下轨、放量下跌、持仓减少。  
- 最佳实践：**基本面定方向，技术面定入场点，资金面判断强弱，策略面控制风险。**


## 5. 实操环节：如何开始交易豆粕期货

---

### （1）开户流程与要求

1. **选择期货公司**  
   - 在中国，需通过中国证监会批准的正规期货公司（大商所会员）。  
   - 建议选择大型、风控与服务完善的公司。  

2. **投资者适当性评估**  
   - 法规要求：申请商品期货账户，需完成期货知识测评。  
   - 部分公司会考核投资经验（如模拟交易记录、股票/基金经验）。  

3. **资金门槛**  
   - 普通商品期货（如豆粕）：开户无硬性资金门槛，一般 0 元起开，但需入金后才能交易。  
   - 特殊品种（如原油、铁矿石）：可能要求 50 万资金+10 笔以上模拟或实盘交易记录。  

4. **入金与保证金**  
   - 将资金转入期货公司专用保证金账户。  
   - 资金实时划转，保证金可随时提取。  

---

### （2）交易规则与成本

- **合约规则（以大连商品交易所豆粕为例）**  
  - 合约规模：**10 吨/手**。  
  - 最小变动价位：**1 元/吨**，即每跳动 = 10 元/手。  
  - 涨跌停幅度：通常为前一交易日结算价的 **±4%**（遇重大行情可调整）。  

- **保证金制度（杠杆效应）**  
  - 保证金比例：约 **5%–8%**（交易所基准，不同公司可能略高）。  
  - 示例：若豆粕价格 3,500 元/吨，一手价值 = 3,500 × 10 = 35,000 元；  
    - 保证金比例 8% → 入金仅需 2,800 元就能交易一手。  
  - **杠杆风险**：盈亏均会被放大，价格波动 1% → 保证金可能波动 10%以上。  

- **手续费**  
  - 按手收取：一般 **1.2–1.5 元/手/双边**（不同公司不同）。  
  - 相比股票佣金，期货手续费更低，但因短线交易频繁，累计成本不可忽视。  

- **强平与爆仓**  
  - 若账户权益 ≤ 维持保证金（约开仓保证金的 75%–80%），系统会强行平仓。  
  - 即使未强平，若极端行情发生，账户可能出现 **穿仓风险**（亏损大于本金）。  

---

### （3）风险提示与建议

- **波动大**  
  - 豆粕价格受国际供需、汇率、气候、政策影响，常出现 **单日数百点波动**。  

- **止损机制**  
  - 新手必须设置止损，如单笔亏损超过保证金的 20% 即平仓。  

- **仓位控制**  
  - 建议初学者最多使用 **20–30%资金**参与期货交易。  
  - 避免“满仓梭哈”，分批建仓，留出追加保证金空间。  

- **模拟交易**  
  - 先在交易软件上开模拟盘，熟悉下单方式、手续费、爆仓规则，再进实盘。  

- **常见误区**  
  - 盲目重仓：高杠杆容易快速爆仓；  
  - 只看技术不看基本面：农产品对政策/天气非常敏感；  
  - 不做止损：小亏不止损，容易演变成大亏。  

---

### 📌 小结
- 开户 **门槛低**，但期货交易 **风险高**。  
- 交易规则清晰：**10 吨/手，保证金约 5–8%，手续费约 1.2–1.5 元/手**。  
- 新手应 **模拟盘+小仓位+严格止损**，逐步熟悉市场后再扩大规模。  


````python

# 模拟保证金和盈亏计算
contract_size = 10  # 每手10吨
price_now = 3500
margin_rate = 0.1  # 10%保证金

margin = price_now * contract_size * margin_rate
print(f"开仓1手豆粕合约所需保证金: {margin:.2f} 元")

# 假设价格上涨200元
price_future = price_now + 200
profit = (price_future - price_now) * contract_size
print(f"若价格上涨200元，1手盈利: {profit:.2f} 元")
````

## 💼 技术面量化策略（黑色系示例：RB/HC/I/J/JM）

本节介绍两种简单、实用、可快速落地的技术面策略：**双均线趋势策略** 与 **布林带突破策略**。二者可单独使用，也可组合成“趋势过滤 + 突破入场”的框架。

---

### 📈 一、双均线趋势策略（Dual MA Crossover）

**思想**  
- 价格具有惯性，短期均线（灵敏）相对长期均线（平滑）的**上穿/下穿**可识别趋势方向。  
- **金叉（短上穿长）**→ 做多；**死叉（短下穿长）**→ 做空。  
- 在单边行情（黑色系常见）中更有效，震荡时易产生假信号。

**参数建议**  
- 短期均线：`MA_short = 20`  
- 长期均线：`MA_long = 60`  
> 不同品种可调优：(10,40)、(30,90) 等，需回测选择。

**交易规则**  
- 开多：`MA_short > MA_long`  
- 开空：`MA_short < MA_long`  
- 平仓：与持仓方向相反的交叉出现，或价格回到成本附近（自定）

**伪代码（Python/Pandas 思路）**
```python
# df['Close'] 为价格（如 RB 主力连续收盘价）
short_window, long_window = 20, 60

df['MA_S'] = df['Close'].rolling(short_window).mean()
df['MA_L'] = df['Close'].rolling(long_window).mean()

df['Signal'] = 0
df.loc[df['MA_S'] > df['MA_L'], 'Signal'] = 1     # 多头
df.loc[df['MA_S'] < df['MA_L'], 'Signal'] = -1    # 空头
df['Ret'] = df['Signal'].shift(1) * df['Close'].pct_change()
