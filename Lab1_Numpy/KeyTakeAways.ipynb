```python
x = np.array([
    [1, 2],
    [3, 4]
])
y = np.array([
    [5, 6],
    [7, 8]
])
v = np.array([9, 10])
w = np.array([11, 12])
```

```python
print(x + y)    # 两个矩阵中的【元素】一一对应进行加减乘除（Elementwise sum, difference, product, division）
```

```python
print(x.dot(y)) # 两个【矩阵】的运算
```

在NumPy的官方文档中，对于 `np.dot`（以及更新的 `@` 运算符）在处理一维数组时的行为有明确的定义：
- 如果两个参数都是一维数组，执行向量内积，结果是一个标量。
- 如果其中一个参数是二维数组（矩阵），另一个是一维数组（向量），则执行矩阵-向量乘法。结果始终是一个一维数组。

```python
print(v.dot(w)) # 处理一维数组的【点积（内积）运算规则】：计算它们对应元素的乘积之和。
```

```python
# 直接使用一维数组，会根据运算的需要，智能地将其“临时”当作行向量或列向量来处理。
print(x.dot(v)) # 【v】被当作列向量
print(v.dot(x)) # 【v】被当作行向量
```

```python
w = np.array([
    [1],
    [2],
    [3]
])
v = np.array([1, 2, 3])

print(w.reshape(-1), "\n")
print(v.reshape(-1, 1), "\n")
print(v.reshape(1, 1, -1))
```

`-1` 的作用：是一个占位符，让 NumPy 自动计算该维度的大小。

1. `arr.reshape(-1)`：快速将任何数组展平为一维。
2. `arr.reshape(-1, 1)`：将一维数组转换为 N行1列 的列向量（二维）。
3. `arr.reshape(1, -1)`：将一维数组转换为 1行N列 的行向量（二维）。

**核心原则：总元素数必须保持不变**

2 和 3 是在机器学习中非常常见的操作

---

广播机制（Broadcasting）
- 允许 NumPy 在不同形状的数组之间进行算术运算。

**核心：广播前后，元素总数相等！！！**

例子：
<img src="./image/Broadcasting.png" alt="image.png" style="width: 500px; height: auto; display: block; margin: 0 auto;">

规则一：让所有数组的维度数（ndim）相同。
- 如果数组的维度数不同，会在形状的 **【前面（左边）】** 不断补 `1`，直到所有数组的维度数相同。

> 例如：
> - 一个 `(3,)` 的向量和一个 `(3, 3)` 的矩阵运算。
> - 向量的形状是 `(3,)`，维度数是 `1`。
> - 矩阵的形状是 `(3, 3)`，维度数是 `2`。
> - 为了让他们维度数相同，NumPy 会将向量的形状补 `1`，变成 `(1, 3)`。

规则二：确保所有维度的大小兼容。

在两个数组的每一个维度上，它们的大小必须满足：
- 相等，或者
- 其中一个是 `1`。

如果某个维度上大小是 `1`，NumPy 会沿着这个维度复制数据，使其大小与另一个数组该维度的大小匹配。

> 接上例：
> - 补完后，向量形状为 `(1, 3)`，矩阵形状为 `(3, 3)`。
> - 检查【第一个维度】：`1` 和 `3` -> 其中一个是 `1`，兼容。NumPy 会将向量沿着第 0 维复制 3 次，变成 `(3, 3)`。
> - 检查【第二个维度】：`3` 和 `3` -> 相等，兼容。
> - 现在两个数组都变成了 `(3, 3)`，可以执行元素对元素的运算。



```python
a = np.array([1, 2, 3])
print(a + 5)
```

```python
a = np.array([              # (2, 3)
    [1, 2, 3],
    [4, 5, 6]
])
b = np.array([10, 20, 30])  # (3, ) -> (1, 3) -> (2, 3)
print(a + b)
```

```python
a = np.array([              # (3, 1)
    [1],                    # (3, 1)
    [2],                    # (3, 1) -> (3, 3)
    [3]
])
# 最终广播：
# [[1, 1, 1],
# [2, 2, 2],
# [3, 3, 3]]

b = np.array([10, 20, 30])  # (3, )  ->  (1, 3)
                            # (1, 3) ->  (3, 3)
                            # (3, 3)
# 最终广播：
# [[10, 20, 30],
# [10, 20, 30],
# [10, 20, 30]]
print(a + b)
```

- 补 1：在形状前面补 `1`，使维度数相同。
- 伸 1：在任何大小为 1 的维度上复制数据，使其大小与另一个数组匹配。

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label = "Sine")
plt.plot(x, y_cos, label = "Cosine")
plt.legend()

plt.xlabel("X axis label")
plt.ylabel("Y axis label")
plt.title("Sine and Cosine")

plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.randint(0, 2, 50)

plt.scatter(x, y, c = colors)
plt.show()
```

```python
x = np.arange(0, 10 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title("Sine")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title("Cosine")
plt.grid()

plt.subplots_adjust(hspace = 0.4)
plt.show()
```
